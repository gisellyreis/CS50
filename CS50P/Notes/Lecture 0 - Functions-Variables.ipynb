* we can nest functions. that is, put one function call, that is to use of a function, inside og the use another function so that the return value of the inner function becomes the argument to or the input to the outer function.
```
  x = int(input("what's x? "))

```

* Floating point or float

  a float is a number with a decimal point. Mathematically is a real number, a number tha has a decimal point in it.
```
  x = float(input("what's x? ")) # 1.2
  y = float(input("what's y? ")) # 3.4
  print(x + y)

  # round the result to the nearest possible integer
  z = round(x + y)
  print(z)
```
* Numeric formatting
```
  x = int(input("what's x? "))
  y = int(input("what's y? "))

  # specify number of digits, second parameter of round()
  z = round(x / y, 2)
  print(z)
```
  or
```
  # specify using an F string how many digits you want to print.
  print(f"{z:.2f}")
```

## Defining functions: def

  def for define. it is used to define, create, invent your own functions, you get to chose the name of the function.

* hello.py
```
  def hello():
    print("hello")

  name = input("what's your name? ")
  hello()
  print(name)
```
  \# passing parameters to my function
```
  def hello(to):
    print("hello,", to)

  name = input("what's your name? ")
  hello(name)
```
  \# "name" is copied to another variable called "to".

  \# set the value of the parameter by default (if not provided by the user) to be equal to "world".
```
  def hello(to="world"):
    print("hello,", to)

  hello()

  name = input("what's your name? ")
  hello(name)
```
  * I have to define any functions at the top of the file or just before calling (using) it.

  ```
    def main():
      name = input("what's your name? ")
      hello(name)

    def hello(to="world"):
      print("hello,", to)

    # we have to actually use or call main()
    main()
  ```
